```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
website_url = 'https://www.tennessee-demographics.com/zip_codes_by_population'
response = requests.get(website_url)

response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
tables = soup.find_all('table', attrs = {'class': 'ranklist'})
```

```python
p1 = pd.read_csv('../data/city_p1.csv')
p2 = pd.read_csv('../data/city_p2.csv')
p3 = pd.read_csv('../data/city_p3.csv')
p4 = pd.read_csv('../data/city_p4.csv')
p5 = pd.read_csv('../data/city_p5.csv')
p6 = pd.read_csv('../data/city_p6.csv')
```

```python
city_pop_and_income = pd.merge(p1, p2,
                              how='outer')
city_pop_and_income = pd.merge(city_pop_and_income, p3,
                              how='outer')
city_pop_and_income = pd.merge(city_pop_and_income, p4,
                              how='outer')
city_pop_and_income = pd.merge(city_pop_and_income, p5,
                              how='outer')
city_pop_and_income = pd.merge(city_pop_and_income, p6,
                              how='outer')
city_pop_and_income.shape
```

```python
city_pop_and_income = pd.read_csv('../data/city_pop_income.csv')
city_pop_and_income.head()
```

```python
city_pop_and_income['location'] = city_pop_and_income['Location (# Zip Codes)'].apply(lambda x: x.split(',')[0])
city_pop_and_income.head()
```

```python
city_pop_and_income = city_pop_and_income[['location', 'Population', 'Avg. Income/Person']]
city_pop_and_income.head()
```

```python
city_pop_and_income.columns =['location', 'population', 'avg_income_pc']
city_pop_and_income.head()
```
